```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn
```

```python
error_curves = [None] * 10
for i in xrange(10):
    error_curves[i] = [line[:-1] for line in open('logs/%d.log'%i)]
    error_curves[i] = [float(x.split(' ')[0]) for x in error_curves[i][4:-1]]
```

```python
plt.figure(figsize=(15,10))
for curve in error_curves:
    plt.plot(curve)
```
